```python
import numpy as np
import time
import matplotlib.pyplot as plt




ROWS = 10
AISLES = 2  # For a 3-4-3 configuration
SEAT_PER_SECTION = [3, 4, 3]
PASSENGERS = sum([x*ROWS for x in SEAT_PER_SECTION])
DELAY = 0.1

class Passenger:
    def __init__(self, seat_row, seat_section):
        self.seat_row = seat_row
        self.seat_section = seat_section
        self.in_aisle = False
        self.boarded = False
        self.aisle_choice = 0 if self.seat_row < ROWS/2 else 1  # Front if seated in front half, else back

aisle_q = [None for _ in range(ROWS*AISLES)]
passenger_list = []

def AssignSeats():
    for i in range(ROWS):
        for j, seats in enumerate(SEAT_PER_SECTION):
            for _ in range(seats):
                passenger_list.append(Passenger(i, j))

def DisplayAisle():
    for i in range(0, len(aisle_q), AISLES):
        print(aisle_q[i:i+AISLES])

def MoveToAisle():
    for i, passenger in enumerate(aisle_q):
        if passenger and passenger.seat_row < i//AISLES:
            aisle_q[i], aisle_q[i-AISLES] = aisle_q[i-AISLES], aisle_q[i]

def Boarding():
    for i, passenger in enumerate(aisle_q):
        if passenger and passenger.seat_row == i//AISLES and not passenger.boarded:
            aisle_q[i] = None
            passenger.boarded = True

def UpdateAisle():
    for i in reversed(range(AISLES, len(aisle_q))):
        if aisle_q[i-AISLES] and not aisle_q[i]:
            aisle_q[i], aisle_q[i-AISLES] = aisle_q[i-AISLES], aisle_q[i]

def EnterAisle():
    # Front entry
    if not aisle_q[0] and passenger_list:
        next_passenger = passenger_list.pop(0)
        if next_passenger.aisle_choice == 0:
            next_passenger.in_aisle = True
            aisle_q[0] = next_passenger
    
    # Back entry
    if not aisle_q[-1] and passenger_list:
        next_passenger = passenger_list.pop()
        if next_passenger.aisle_choice == 1:
            next_passenger.in_aisle = True
            aisle_q[-1] = next_passenger

AssignSeats()
np.random.shuffle(passenger_list)

steps = 0
while len([p for p in passenger_list if not p.boarded]) > 0:
    EnterAisle()
    Boarding()
    MoveToAisle()
    UpdateAisle()
    DisplayAisle()
    steps += 1
    time.sleep(DELAY)

print(f"Boarding completed in {steps} steps.")
```
